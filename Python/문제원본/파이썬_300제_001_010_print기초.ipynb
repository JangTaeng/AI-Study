### 005 print 탭과 줄바꿈

다음 코드를 실행해보고 `\t`와 `\n`의 역할을 설명해보세요.
```python
print("안녕하세요.\n만나서\t\t반갑습니다.")
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<code>\n</code>은 줄바꿈(개행)을 의미합니다.</br>
<code>\t</code>는 탭(Tab) 간격을 의미합니다. 키보드의 Tab 키를 누른 것과 같은 효과입니다.</br>
<pre>
# 출력 결과:
# 안녕하세요.
# 만나서		반갑습니다.
print("안녕하세요.\n만나서\t\t반갑습니다.")
</pre>
</details>

````python
# 직접 실행해서 결과를 확인해보세요
print("안녕하세요.\n만나서\t\t반갑습니다.")
````

### 009 print 줄바꿈

다음 코드를 수정하여 줄바꿈 없이 한 줄에 출력하세요. (힌트: `end=''`)</br>
print 함수는 두 번 사용합니다.
```python
print("first"); print("second")
```
```
실행 예:
firstsecond
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
print 함수는 기본적으로 출력 후 줄바꿈을 합니다. <code>end=""</code>를 사용하면 줄바꿈 대신 빈 문자열로 끝나게 됩니다.</br>
세미콜론(;)은 한 줄에 여러 개의 명령을 작성할 때 사용합니다.</br>
<pre>
print("first", end=""); print("second")
</pre>
</details>

````python
# 아래 코드를 수정해보세요
print("first"); print("second")
````
