### 280 입출금 내역 기록 메서드

```
실행 예:
k.deposit(100)
k.deposit(200)
k.deposit_history()
# 100
# 200
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
생성자에서 빈 로그 리스트를 초기화하고, 입출금 시 append합니다.</br>
<pre>
def __init__(self, ...):
    self.deposit_log = []
    self.withdraw_log = []
    # ...

def deposit(self, amount):
    if amount >= 1:
        self.deposit_log.append(amount)
        self.balance += amount

def withdraw(self, amount):
    if self.balance > amount:
        self.withdraw_log.append(amount)
        self.balance -= amount

def deposit_history(self):
    for amount in self.deposit_log:
        print(amount)

def withdraw_history(self):
    for amount in self.withdraw_log:
        print(amount)
</pre>
