### 531 클래스 - __contains__ 구현

in 연산자를 지원하도록 __contains__를 구현하세요.
```python
wl = Whitelist(['admin','user','guest'])
'admin' in wl  # True
'hacker' in wl  # False
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class Whitelist:
    def __init__(self,items): self.items=set(items)
    def __contains__(self,item): return item in self.items
    def add(self,item): self.items.add(item)
    def remove(self,item): self.items.discard(item)
    def __repr__(self): return f'Whitelist({self.items})'

wl=Whitelist(['admin','user','guest'])
print('admin' in wl)   # True
print('hacker' in wl)  # False
wl.add('hacker')
print('hacker' in wl)  # True

### 532 클래스 - __iter__ + __getitem__

인덱싱과 반복 순회를 모두 지원하는 Matrix 클래스를 구현하세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class Matrix:
    def __init__(self,data): self.data=data
    def __getitem__(self,idx):
        if isinstance(idx,tuple): return self.data[idx[0]][idx[1]]
        return self.data[idx]
    def __iter__(self):
        for row in self.data:
            for val in row: yield val
    def __repr__(self):
        return '\n'.join(str(row) for row in self.data)

m=Matrix([[1,2,3],[4,5,6],[7,8,9]])
print(m[1,2])        # 6 (1행 2열)
print(m[0])          # [1,2,3]
print(list(m))       # 모든 원소 1차원
</pre>

### 534 클래스 - 프로퍼티로 유효성 검사

@property와 setter를 이용해 입력 유효성을 검사하는 클래스를 구현하세요.
```python
p = Person('김파이', 25)
p.age = -5  # ValueError 발생
p.name = ''  # ValueError 발생
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class Person:
    def __init__(self,name,age):
        self.name=name  # setter 호출
        self.age=age
    @property
    def name(self): return self._name
    @name.setter
    def name(self,v):
        if not v or not isinstance(v,str): raise ValueError('이름은 비어있을 수 없습니다')
        self._name=v
    @property
    def age(self): return self._age
    @age.setter
    def age(self,v):
        if not isinstance(v,int) or v<0 or v>150: raise ValueError(f'나이 오류: {v}')
        self._age=v
    def __repr__(self): return f'Person({self._name},{self._age})'

p=Person('김파이',25)
print(p)
try: p.age=-5
except ValueError as e: print(e)
try: p.name=''
except ValueError as e: print(e)

### 535 클래스 - 믹스인(Mixin) 다중 활용

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class LogMixin:
    def log(self,msg): print(f'[LOG] {self.__class__.__name__}: {msg}')

class ValidateMixin:
    def validate(self,value,type_):
        if not isinstance(value,type_): raise TypeError(f'{value}는 {type_.__name__} 타입이어야 합니다')
        return True

class SerializeMixin:
    def to_dict(self): return {k:v for k,v in self.__dict__.items() if not k.startswith('_')}

class User(LogMixin,ValidateMixin,SerializeMixin):
    def __init__(self,name,age):
        self.validate(name,str)
        self.validate(age,int)
        self.name=name; self.age=age
        self.log(f'{name} 생성됨')

u=User('김파이',25)

try: User('이파이','삼십')

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class Multiplier:
    def __init__(self,factor): self.factor=factor; self.call_count=0
    def __call__(self,value):
        self.call_count+=1
        return value*self.factor
    def __repr__(self): return f'Multiplier(x{self.factor}, {self.call_count}회 호출)'

double=Multiplier(2)

### 537 클래스 - 체이닝 빌더 패턴

메서드 체이닝으로 SQL WHERE 조건을 빌드하는 클래스를 구현하세요.
```python
q = Query('users').select('name','age').where(city='서울').where(age_gt=20).limit(10)
q.build()  # 'SELECT name,age FROM users WHERE city=서울 AND age>20 LIMIT 10'
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class Query:
    def __init__(self,table):
        self._table=table; self._cols=['*']; self._conds=[]; self._lim=None
    def select(self,*cols): self._cols=list(cols); return self
    def where(self,**kwargs):
        for k,v in kwargs.items():
            if k.endswith('_gt'): self._conds.append(f'{k[:-3]}>{v}')
            elif k.endswith('_lt'): self._conds.append(f'{k[:-3]}<{v}')
            else: self._conds.append(f'{k}={v}')
        return self
    def limit(self,n): self._lim=n; return self
    def build(self):
        sql=f"SELECT {','.join(self._cols)} FROM {self._table}"
        if self._conds: sql+=' WHERE '+' AND '.join(self._conds)
        if self._lim: sql+=f' LIMIT {self._lim}'
        return sql

q=Query('users').select('name','age').where(city='서울').where(age_gt=20).limit(10)
print(q.build())

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self): pass
    @abstractmethod
    def perimeter(self): pass
    def describe(self): print(f'{self.__class__.__name__}: 넓이={self.area():.2f}, 둘레={self.perimeter():.2f}')

class Circle(Shape):
    def __init__(self,r): self.r=r
    def area(self): return 3.14159*self.r**2
    def perimeter(self): return 2*3.14159*self.r

class Rectangle(Shape):
    def __init__(self,w,h): self.w=w; self.h=h
    def area(self): return self.w*self.h
    def perimeter(self): return 2*(self.w+self.h)

for shape in [Circle(5), Rectangle(4,6)]:
    shape.describe()

try:
    class BadShape(Shape): pass
    BadShape()  # 오류

### 540 클래스 종합 - 제품 관리 시스템

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class Product:
    _count=0
    def __init__(self,name,price,stock,category):
        Product._count+=1
        self.id=Product._count
        self.name=name; self.price=price; self.stock=stock; self.category=category
    def __repr__(self): return f'[{self.id}]{self.name}({self.price:,}원,재고:{self.stock})'

class ProductManager:
    def __init__(self): self._products={}  # id → Product
    def add(self,name,price,stock,category='기타'):
        p=Product(name,price,stock,category)
        self._products[p.id]=p; return p
    def get(self,id_): return self._products.get(id_)
    def search(self,keyword): return [p for p in self._products.values() if keyword in p.name]
    def by_category(self):
        result={}
        for p in self._products.values(): result.setdefault(p.category,[]).append(p)
        return result
    def top_expensive(self,n=3): return sorted(self._products.values(),key=lambda p:-p.price)[:n]
    def low_stock(self,threshold=10): return [p for p in self._products.values() if p.stock<=threshold]
    def report(self):
        print(f'=== 제품 현황 (총 {len(self._products)}개) ===')
        for cat,products in self.by_category().items(): print(f'  {cat}: {products}')
        print(f'  고가 TOP3: {self.top_expensive()}')
        print(f'  재고부족: {self.low_stock()}')

pm=ProductManager()
pm.add('파이썬 교재',25000,50,'도서')
pm.add('노트북',1500000,8,'전자제품')
pm.add('마우스',35000,5,'전자제품')
pm.add('파이썬 강의',99000,999,'교육')
pm.add('키보드',85000,3,'전자제품')
pm.report()
