### 152 for문 + 조건문 (3의 배수 출력)

for문을 사용해서 3의 배수만을 출력하라.
```python

```
```
실행 예:
```

### 157 for문 + 첫 글자 대문자 변환

이름의 첫 글자를 대문자로 변경해서 출력하라.
```python
리스트 = ['dog', 'cat', 'parrot']
```
```
실행 예:
Dog
Cat
Parrot
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
인덱싱으로 첫 글자를 대문자로 바꾼 뒤 나머지 문자열과 이어붙입니다.</br>
<pre>
리스트 = ['dog', 'cat', 'parrot']
for 변수 in 리스트:
    print(변수[0].upper() + 변수[1:])

# 또는 capitalize() 메서드 활용:

    print(변수.capitalize())
</pre>
</details>

### 158 for문 + split (확장자 제거)

파일 이름이 저장된 리스트에서 확장자를 제거하고 파일 이름만 출력하라.
```python
리스트 = ['hello.py', 'ex01.py', 'intro.hwp']
```
```
실행 예:
hello
ex01
intro
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
split('.')으로 파일명과 확장자를 분리하고, 인덱스 0(파일명)만 출력합니다.</br>
<pre>
리스트 = ['hello.py', 'ex01.py', 'intro.hwp']
for 변수 in 리스트:
    split = 변수.split(".")
    print(split[0])
</pre>
</details>

````python
리스트 = ['hello.py', 'ex01.py', 'intro.hwp']
# 여기에 코드를 작성하세요
````

### 159 for문 + split + 조건 (.h 파일 출력)

<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>

for 변수 in 리스트:

### 160 for문 + split + or 조건 (.h 또는 .c 파일)

파일 이름이 저장된 리스트에서 확장자가 .h나 .c인 파일을 출력하라.
```python
리스트 = ['intra.h', 'intra.c', 'define.h', 'run.py']
```
```
실행 예:
intra.h
intra.c
define.h
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
or 연산자로 두 개의 확장자를 동시에 검사합니다.</br>
<pre>
리스트 = ['intra.h', 'intra.c', 'define.h', 'run.py']
for 변수 in 리스트:
    split = 변수.split(".")
    if (split[1] == "h") or (split[1] == "c"):
        print(변수)

# endswith 활용:
for 변수 in 리스트:
    if 변수.endswith((".h", ".c")):
        print(변수)
</pre>
</details>

````python
리스트 = ['intra.h', 'intra.c', 'define.h', 'run.py']
# 여기에 코드를 작성하세요
````
