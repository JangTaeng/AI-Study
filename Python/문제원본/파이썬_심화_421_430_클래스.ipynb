class Model:
    _db=[]
    @classmethod
    def create(cls,**kwargs):
        obj=cls.__new__(cls)
        for k,v in kwargs.items(): setattr(obj,k,v)
        cls._db.append(obj)
        return obj
    @classmethod
    def all(cls): return cls._db
    @classmethod
    def filter(cls,**kwargs):
        return [o for o in cls._db if all(getattr(o,k,None)==v for k,v in kwargs.items())]

class User(Model): pass

User.create(name='김파이',age=25,city='서울')
User.create(name='이파이',age=30,city='부산')
User.create(name='박파이',age=25,city='서울')

class Calculator:
    def __init__(self): self.result=0; self.history=[]
    def _op(self,op,val):
        old=self.result
        if op=='+': self.result+=val
        elif op=='-': self.result-=val
        elif op=='*': self.result*=val
        elif op=='/': self.result/=val if val!=0 else (print('0으로 나누기 불가') or val)
        self.history.append(f'{old} {op} {val} = {self.result}')
        return self
    def add(self,v): return self._op('+',v)
    def sub(self,v): return self._op('-',v)
    def mul(self,v): return self._op('*',v)
    def div(self,v): return self._op('/',v)
    def reset(self): self.result=0; return self
    def show_history(self): [print(h) for h in self.history]

class Config:
    def __init__(self): self._data={}
    def set(self,key,value):
        keys=key.split('.')
        d=self._data
        for k in keys[:-1]: d=d.setdefault(k,{})
        d[keys[-1]]=value
    def get(self,key,default=None):
        keys=key.split('.')
        d=self._data
        for k in keys:
            if not isinstance(d,dict) or k not in d: return default
            d=d[k]
        return d

cfg.set('database.host','localhost')

<summary>정답 확인 (클릭!)</summary>
<pre>
class EventBus:
    def __init__(self): self._handlers={}
    def on(self,event,handler):
        self._handlers.setdefault(event,[]).append(handler)
    def emit(self,event,*args,**kwargs):
        for h in self._handlers.get(event,[]): h(*args,**kwargs)

bus=EventBus()
bus.on('login', lambda user: print(f'{user} 로그인'))
bus.on('login', lambda user: print(f'환영합니다, {user}님!'))
bus.on('logout', lambda user: print(f'{user} 로그아웃'))

import random, datetime

class Transaction:
    def __init__(self,type_,amount,balance):
        self.type=type_; self.amount=amount; self.balance=balance
        self.date=datetime.datetime.now().strftime('%Y-%m-%d %H:%M')
    def __repr__(self): return f'[{self.date}] {self.type} {self.amount:,}원 (잔고:{self.balance:,}원)'

class Account:
    def __init__(self,name,balance):
        self.name=name; self.balance=balance
        self.transactions=[]
    def deposit(self,amount):
        self.balance+=amount
        self.transactions.append(Transaction('입금',amount,self.balance))
    def withdraw(self,amount):
        if self.balance>=amount:
            self.balance-=amount
            self.transactions.append(Transaction('출금',amount,self.balance))
        else: print('잔고 부족')
    def transfer(self,target,amount):

            self.withdraw(amount); target.deposit(amount)
            print(f'{self.name}→{target.name} {amount:,}원 송금 완료')
    def show_history(self): [print(t) for t in self.transactions]
