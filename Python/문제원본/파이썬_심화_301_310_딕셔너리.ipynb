### 303 setdefault() 메서드

setdefault()를 사용해서 키가 없을 때만 기본값을 설정하세요.
```python
cart = {'사과': 3}
# '바나나' 키가 없으면 0으로 초기화 후 2 추가
# '사과' 키가 이미 있으면 변경하지 않음
```
```
실행 예:
{'사과': 3, '바나나': 2}
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
cart = {'사과': 3}

cart.setdefault('바나나', 0)
cart['바나나'] += 2

cart.setdefault('사과', 0)  # 이미 있으므로 변경 없음

print(cart)  # {'사과': 3, '바나나': 2}
</pre>
</details>

````python
cart = {'사과': 3}
# 여기에 코드를 작성하세요
````

### 304 딕셔너리 병합 (update)

```python
d1 = {'a': 1, 'b': 2}
d2 = {'b': 99, 'c': 3}
# 결과: {'a': 1, 'b': 99, 'c': 3}
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>

d2 = {'b': 99, 'c': 3}

# 방법 1: update()

print(d1)  # {'a': 1, 'b': 99, 'c': 3}

# 방법 2: ** 언패킹 (원본 유지)
merged = {**d1, **d2}
print(merged)

# 방법 3: Python 3.9+
# merged = d1 | d2
</pre>
</details>

````python
d1 = {'a': 1, 'b': 2}
d2 = {'b': 99, 'c': 3}
# 여기에 코드를 작성하세요
````

### 310 딕셔너리 키/값 삭제

딕셔너리에서 키를 삭제하는 3가지 방법을 사용해보세요.
```python
d = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
# 'a'는 del로, 'b'는 pop()으로, 'c'는 popitem()으로 삭제
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
d = {'a': 1, 'b': 2, 'c': 3, 'd': 4}

# 방법 1: del
del d['a']

# 방법 2: pop() - 삭제된 값 반환
val = d.pop('b')
print('삭제된 값:', val)  # 2

# 방법 3: popitem() - 마지막 항목 삭제 및 반환 (Python 3.7+)
item = d.popitem()
print('삭제된 항목:', item)

print(d)
</pre>
</details>

````python
d = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
# 여기에 코드를 작성하세요
````
