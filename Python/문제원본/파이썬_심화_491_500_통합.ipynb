import heapq, datetime
class Task:
    def __init__(self,name,priority,func):
        self.name=name; self.priority=priority; self.func=func
    def __lt__(self,other): return self.priority<other.priority
class Scheduler:
    def __init__(self): self.queue=[]
    def add(self,name,priority,func=None):
        heapq.heappush(self.queue,Task(name,priority,func or (lambda:print(f'{name} 실행'))))
    def run_next(self):
        if self.queue:
            task=heapq.heappop(self.queue)
            print(f'[우선순위:{task.priority}] {task.name} 실행')
            task.func()
    def run_all(self):
        while self.queue: self.run_next()

class Trackable:
    def __init__(self): object.__setattr__(self,'_history',{})
    def __setattr__(self,name,value):
        if not name.startswith('_'):
            hist=object.__getattribute__(self,'_history')
            hist.setdefault(name,[]).append(value)
        object.__setattr__(self,name,value)
    def history(self,name): return object.__getattribute__(self,'_history').get(name,[])
    def rollback(self,name):
        hist=object.__getattribute__(self,'_history').get(name,[])
        if len(hist)>1: hist.pop(); object.__setattr__(self,name,hist[-1])

class Product(Trackable):
    def __init__(self,name,price): super().__init__(); self.name=name; self.price=price

p=Product('사과',3000)
p.price=3500; p.price=4000
print(p.history('price'))   # [3000,3500,4000]
p.rollback('price')
print(p.price)              # 3500

class VersionedDict:
    def __init__(self): self._current={}; self._versions=[{}]
    def __setitem__(self,k,v):
        self._current[k]=v
        self._versions.append(dict(self._current))
    def __getitem__(self,k): return self._current[k]
    def __repr__(self): return str(self._current)
    def rollback(self,steps=1):
        if len(self._versions)>steps:
            self._current=dict(self._versions[-steps-1])
            self._versions=self._versions[:-steps]
    def version_count(self): return len(self._versions)

import random
class ABTest:
    def __init__(self,name,variants,weights=None):
        self.name=name; self.variants=variants
        self.weights=weights or [1/len(variants)]*len(variants)
        self.results={v:{'views':0,'conversions':0} for v in variants}
    def assign(self): return random.choices(self.variants,weights=self.weights)[0]
    def record(self,variant,converted):
        self.results[variant]['views']+=1
        if converted: self.results[variant]['conversions']+=1
    def report(self):
        print(f'=== {self.name} A/B 테스트 결과 ===')
        for v,r in self.results.items():
            rate=r['conversions']/r['views']*100 if r['views'] else 0
            print(f'{v}: {r["views"]}회 노출, {r["conversions"]}회 전환 ({rate:.1f}%)')

test=ABTest('버튼 색상',['빨강','파랑'],[0.5,0.5])
for _ in range(100):
    v=test.assign()
    converted=random.random()<(0.15 if v=='빨강' else 0.12)
    test.record(v,converted)
test.report()

### 497 통합 - 간단한 그래프 탐색 (BFS)

딕셔너리 그래프에서 BFS로 최단 경로를 찾으세요.
```python
graph={'A':['B','C'],'B':['D'],'C':['D','E'],'D':['F'],'E':['F']}
# A→F 최단경로
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
from collections import deque
graph={'A':['B','C'],'B':['D'],'C':['D','E'],'D':['F'],'E':['F'],'F':[]}
def bfs(graph,start,end):
    queue=deque([[start]])
    visited={start}
    while queue:
        path=queue.popleft()
        node=path[-1]
        if node==end: return path
        for neighbor in graph.get(node,[]):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(path+[neighbor])
    return None
print('최단경로:', bfs(graph,'A','F'))

class Stream:
    def __init__(self,data): self._data=list(data)
    def filter(self,f): return Stream(x for x in self._data if f(x))
    def map(self,f): return Stream(f(x) for x in self._data)
    def sort(self,key=None,reverse=False): return Stream(sorted(self._data,key=key,reverse=reverse))
    def group_by(self,f):
        result={}
        for x in self._data: result.setdefault(f(x),[]).append(x)
        return result
    def count(self): return len(self._data)
    def first(self,n=1): return self._data[:n] if n>1 else self._data[0] if self._data else None
    def to_list(self): return self._data

def run_tests(func,test_cases):
    passed=failed=0
    for case in test_cases:
        args=case.get('args',[])
        kwargs=case.get('kwargs',{})
        expected=case['expected']
        try:
            result=func(*args,**kwargs)
            if result==expected: print(f'✅ PASS: {func.__name__}{tuple(args)} = {result}'); passed+=1
            else: print(f'❌ FAIL: {func.__name__}{tuple(args)} = {result}, 기대:{expected}'); failed+=1
        except Exception as e: print(f'❌ ERROR: {e}'); failed+=1
    print(f'결과: {passed}/{passed+failed} 통과')

def is_palindrome(s): return s==s[::-1]

class Book:
    def __init__(self,isbn,title,author,year):
        self.isbn=isbn; self.title=title; self.author=author; self.year=year
        self.is_available=True; self.borrower=None
    def __repr__(self): return f'[{"가능" if self.is_available else self.borrower}] {self.title}({self.author})'

class Library:
    def __init__(self,name): self.name=name; self.books={}; self.members={}
    def add_book(self,isbn,title,author,year):
        self.books[isbn]=Book(isbn,title,author,year)
    def register_member(self,member_id,name): self.members[member_id]={'name':name,'borrowed':[]}
    def borrow(self,member_id,isbn):
        if member_id not in self.members: return print('미등록 회원')
        if isbn not in self.books: return print('없는 도서')
        book=self.books[isbn]
        if not book.is_available: return print(f'{book.title}: 대출 중')
        book.is_available=False; book.borrower=self.members[member_id]['name']
        self.members[member_id]['borrowed'].append(isbn)
        print(f"{self.members[member_id]['name']}님이 [{book.title}] 대출")
    def return_book(self,member_id,isbn):
        if isbn in self.members[member_id]['borrowed']:
            self.books[isbn].is_available=True; self.books[isbn].borrower=None
            self.members[member_id]['borrowed'].remove(isbn)
            print(f"{self.members[member_id]['name']}님이 [{self.books[isbn].title}] 반납")
    def search(self,keyword): return [b for b in self.books.values() if keyword in b.title or keyword in b.author]
    def status(self):
        print(f'=== {self.name} 현황 ===')
        available=sum(1 for b in self.books.values() if b.is_available)
        print(f'총 {len(self.books)}권 (가능:{available}, 대출:{len(self.books)-available})')
        for b in self.books.values(): print(' ',b)

lib=Library('파이썬도서관')
for isbn,t,a,y in [('001','파이썬 기초','김파이',2020),('002','알고리즘','이파이',2021),('003','데이터 분석','박파이',2022)]:
