### 402 클래스 특수 메서드 __repr__

<pre>

    def __init__(self,name,age):

        return f'이름: {self.name}, 나이: {self.age}'

p=Person('김파이',25)
print(repr(p))  # Person('김파이', 25)

<details>
<summary>정답 확인 (클릭!)</summary>

class Score:

### 406 @property 데코레이터

@property를 사용해서 섭씨(celsius)와 화씨(fahrenheit)를 자동 변환하는 클래스를 만드세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class Temperature:
    def __init__(self,celsius):
        self._celsius=celsius
    @property
    def celsius(self): return self._celsius
    @celsius.setter
    def celsius(self,val): self._celsius=val
    @property
    def fahrenheit(self): return self._celsius*9/5+32

t=Temperature(0)
print(t.celsius)     # 0
print(t.fahrenheit)  # 32.0
t.celsius=100
print(t.fahrenheit)  # 212.0
</pre>

### 407 클래스 메서드 vs 정적 메서드

@classmethod와 @staticmethod의 차이를 코드로 확인하세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class Calculator:
    tax_rate=0.1  # 클래스 변수

    def __init__(self,price): self.price=price

    def get_price_with_tax(self):
        return self.price*(1+Calculator.tax_rate)

    @classmethod
    def set_tax(cls,rate): cls.tax_rate=rate  # 클래스 변수 변경

    @staticmethod
    def add(a,b): return a+b  # 클래스/인스턴스와 무관

c=Calculator(1000)
print(c.get_price_with_tax())  # 1100.0
Calculator.set_tax(0.2)
print(c.get_price_with_tax())  # 1200.0
print(Calculator.add(3,4))     # 7

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
from abc import ABC, abstractmethod

class Animal(ABC):
    def __init__(self,name): self.name=name
    @abstractmethod
    def sound(self): pass
    def info(self): print(f'{self.name}: {self.sound()}')

class Dog(Animal):
    def sound(self): return '멍멍'

class Cat(Animal):
    def sound(self): return '야옹'

d=Dog('바둑이'); c=Cat('나비')
d.info()

### 410 클래스 데코레이터 패턴

클래스를 사용한 데코레이터 패턴을 구현하세요. 함수 호출 횟수를 카운트하는 클래스를 만드세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class CallCounter:
    def __init__(self,func):
        self.func=func
        self.count=0
    def __call__(self,*args,**kwargs):
        self.count+=1
        print(f'{self.func.__name__} 호출 {self.count}회')
        return self.func(*args,**kwargs)

@CallCounter
def greet(name): print(f'안녕, {name}!')

greet('민수')
greet('지영')
greet('철수')
print('총 호출:', greet.count)
