### 375 리스트 중복 쌍 찾기

리스트에서 합이 특정 값이 되는 모든 쌍을 찾으세요.
```python
nums=[1,5,3,4,2]; target=5
# 결과: [(1,4),(3,2)]
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
nums=[1,5,3,4,2]; target=5
seen=set(); pairs=[]
for n in nums:
    comp=target-n
    if comp in seen: pairs.append((min(n,comp),max(n,comp)))
    seen.add(n)
print(list(set(pairs)))
</pre>
</details>

### 380 리스트 원소 좌우 곱

각 위치에서 해당 원소를 제외한 나머지의 곱을 구하세요.
```python
nums=[1,2,3,4]
# 결과: [24,12,8,6] (2*3*4, 1*3*4, 1*2*4, 1*2*3)
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
nums=[1,2,3,4]
n=len(nums)
left=[1]*n; right=[1]*n
for i in range(1,n): left[i]=left[i-1]*nums[i-1]
for i in range(n-2,-1,-1): right[i]=right[i+1]*nums[i+1]
print([left[i]*right[i] for i in range(n)])
</pre>
