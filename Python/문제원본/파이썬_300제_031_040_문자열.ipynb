</details>

### 035 문자열 출력 (% formatting)

변수에 다음과 같이 문자열과 정수가 바인딩되어 있을 때 `% formatting`을 사용해서 출력해보세요.
```python
name1 = "김민수"
age1 = 10
name2 = "이철희"
age2 = 13
```
```
실행 예:
이름: 김민수 나이: 10
이름: 이철희 나이: 13
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<code>%s</code>는 문자열, <code>%d</code>는 정수 타입의 값을 출력합니다.</br>
<pre>
name1 = "김민수"
age1 = 10
name2 = "이철희"
age2 = 13
print("이름: %s 나이: %d" % (name1, age1))
print("이름: %s 나이: %d" % (name2, age2))
</pre>
</details>

````python
name1 = "김민수"
age1 = 10
name2 = "이철희"
age2 = 13
# 여기에 코드를 작성하세요
````

### 036 문자열 출력 (format 메서드)

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
format 메서드는 타입과 상관없이 값이 출력될 위치에 <code>{}</code>를 사용합니다.</br>

</pre>
</details>

````python
name1 = "김민수"
age1 = 10
name2 = "이철희"
age2 = 13
# 여기에 코드를 작성하세요
````

</pre>
