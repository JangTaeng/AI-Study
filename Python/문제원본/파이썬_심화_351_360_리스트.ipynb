### 358 리스트를 이용한 스택 구현

리스트로 스택을 구현하고 push, pop, peek, is_empty를 사용해보세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class Stack:
    def __init__(self): self.data=[]
    def push(self,v): self.data.append(v)
    def pop(self): return self.data.pop() if self.data else None
    def peek(self): return self.data[-1] if self.data else None
    def is_empty(self): return len(self.data)==0

s=Stack()
s.push(1); s.push(2); s.push(3)
print(s.peek())   # 3
print(s.pop())    # 3
print(s.is_empty()) # False
</pre>
</details>

### 359 리스트를 이용한 큐 구현

collections.deque를 사용해 큐(FIFO)를 구현하세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
from collections import deque
class Queue:
    def __init__(self): self.data=deque()
    def enqueue(self,v): self.data.append(v)
    def dequeue(self): return self.data.popleft() if self.data else None
    def peek(self): return self.data[0] if self.data else None

q=Queue()
q.enqueue('A'); q.enqueue('B'); q.enqueue('C')
print(q.dequeue())  # A
print(q.peek())     # B
</pre>
</details>
