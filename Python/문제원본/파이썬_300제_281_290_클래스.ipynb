### 284 super() - 부모 생성자 호출

다음 코드가 동작하도록 자전차 클래스를 정의하세요. 단 자전차 클래스는 차 클래스를 상속받습니다.
```python
bicycle = 자전차(2, 100, "시마노")
print(bicycle.구동계)  # 시마노
print(bicycle.바퀴)    # 2
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
super().__init__()으로 부모 클래스의 생성자를 호출합니다.</br>
<pre>
class 차:
    def __init__(self, 바퀴, 가격):
        self.바퀴 = 바퀴
        self.가격 = 가격

class 자전차(차):
    def __init__(self, 바퀴, 가격, 구동계):
        super().__init__(바퀴, 가격)  # 부모 생성자 호출
        self.구동계 = 구동계

bicycle = 자전차(2, 100, "시마노")
print(bicycle.구동계)

### 287 메서드 오버라이딩 + super() 호출

자전차의 정보() 메서드로 구동계 정보까지 출력하도록 수정해보세요.
```python
bicycle = 자전차(2, 100, "시마노")
bicycle.정보()
# 바퀴수  2
# 가격  100
# 구동계  시마노
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
super().정보()로 부모의 정보() 메서드를 먼저 호출한 후 추가 내용을 출력합니다.</br>
<pre>
class 자전차(차):
    def __init__(self, 바퀴, 가격, 구동계):
        super().__init__(바퀴, 가격)
        self.구동계 = 구동계

    def 정보(self):
        super().정보()  # 부모의 정보() 호출
        print("구동계 ", self.구동계)

bicycle = 자전차(2, 100, "시마노")

### 288 메서드 오버라이딩 - 결과 예측

다음 코드의 실행 결과를 예상해보세요.
```python
class 부모:
    def 호출(self):
        print("부모호출")

class 자식(부모):
    def 호출(self):
        print("자식호출")

```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>

<pre>
# 결과:

</pre>
</details>

````python
class 부모:
    def 호출(self):
        print("부모호출")

class 자식(부모):
    def 호출(self):
        print("자식호출")

나 = 자식()
나.호출()
````

### 290 super().__init__() - 결과 예측

다음 코드의 실행 결과를 예상해보세요.
```python
class 부모:
    def __init__(self):
        print("부모생성")

class 자식(부모):
    def __init__(self):
        print("자식생성")
        super().__init__()

나 = 자식()
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
super().__init__()으로 부모 생성자를 명시적으로 호출합니다.</br>
<pre>
# 결과:
# 자식생성
# 부모생성
</pre>
</details>

````python
class 부모:
    def __init__(self):
        print("부모생성")

class 자식(부모):
    def __init__(self):
        print("자식생성")
        super().__init__()

나 = 자식()
````
