config={}
current_section=None
for line in config_str.strip().split('\n'):
    line=line.strip()
    if line.startswith('['):
        current_section=line[1:-1]

### 470 딕셔너리 - 중첩 딕셔너리 검색

중첩 딕셔너리에서 특정 값을 가진 모든 키 경로를 찾으세요.
```python
data={'a':1,'b':{'c':2,'d':{'e':3,'f':2}}}
# 값이 2인 경로: ['b.c','b.d.f']
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
data={'a':1,'b':{'c':2,'d':{'e':3,'f':2}}}
def find_paths(d,target,path=''):
    results=[]
    for k,v in d.items():
        cur=f'{path}.{k}' if path else k
        if v==target: results.append(cur)
        elif isinstance(v,dict): results.extend(find_paths(v,target,cur))
    return results
print(find_paths(data,2))
</pre>
</details>
