### 098 zip으로 딕셔너리 생성

```python

```
```
실행 예시:
{'apple': 300, 'pear': 250, 'peach': 400}
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<code>zip()</code>으로 두 시퀀스를 묶고, <code>dict()</code>로 딕셔너리를 생성합니다.</br>
<pre>
keys = ("apple", "pear", "peach")
vals = (300, 250, 400)
result = dict(zip(keys, vals))
print(result)
</pre>
</details>

````python
keys = ("apple", "pear", "peach")
vals = (300, 250, 400)
# 여기에 코드를 작성하세요
````

### 100 딕셔너리 in 연산자

'메로나'가 icecream 딕셔너리에 존재하는지 확인하세요.
```python
icecream = {'탱크보이': 1200, '폴라포': 1200, '빵빠레': 1800, '월드콘': 1500, '메로나': 1000}
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<code>in</code> 연산자는 딕셔너리에서 키의 존재 여부를 True/False로 반환합니다.</br>
<pre>
icecream = {'탱크보이': 1200, '폴라포': 1200, '빵빠레': 1800, '월드콘': 1500, '메로나': 1000}
print('메로나' in icecream)   # True
print('누가바' in icecream)   # False
</pre>
</details>

````python
icecream = {'탱크보이': 1200, '폴라포': 1200, '빵빠레': 1800, '월드콘': 1500, '메로나': 1000}
# 여기에 코드를 작성하세요
````
