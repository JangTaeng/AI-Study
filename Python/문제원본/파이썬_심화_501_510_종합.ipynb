class Registry:
    def __init__(self): self._store={}; self._next_id=1
    def add(self,name,**attrs):
        id_=self._next_id
        self._store[id_]={'id':id_,'name':name,**attrs}
        self._next_id+=1
        return id_
    def get(self,id_): return self._store.get(id_)
    def all(self): return list(self._store.values())
    def delete(self,id_): return self._store.pop(id_,None) is not None

reg=Registry()
reg.add('사과',price=3000,stock=100)
reg.add('배',price=5000,stock=50)
reg.add('포도',price=8000,stock=30)
print(reg.get(2))
reg.delete(1)
print(reg.all())

### 502 리스트 - 슬라이딩 윈도우 최소값

크기 k인 슬라이딩 윈도우에서 최솟값을 효율적으로 구하세요.
```python
nums=[1,3,-1,-3,5,3,6,7]; k=3
# 결과: [-1,-3,-3,-3,3,3]
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
from collections import deque
nums=[1,3,-1,-3,5,3,6,7]; k=3
result=[]; dq=deque()
for i,n in enumerate(nums):
    while dq and dq[0]<i-k+1: dq.popleft()
    while dq and nums[dq[-1]]>=n: dq.pop()
    dq.append(i)
    if i>=k-1: result.append(nums[dq[0]])
print(result)
</pre>

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
nums=sorted([-1,0,1,2,-1,-4])
result=[]
for i in range(len(nums)-2):
    if i>0 and nums[i]==nums[i-1]: continue
    l,r=i+1,len(nums)-1
    while l<r:
        s=nums[i]+nums[l]+nums[r]
        if s==0: result.append((nums[i],nums[l],nums[r])); l+=1; r-=1
        elif s<0: l+=1
        else: r-=1
print(result)

### 508 리스트 - 최장 공통 부분 수열(LCS)

두 리스트의 최장 공통 부분 수열(LCS)의 길이를 구하세요.
```python
a=[1,3,4,5,2]; b=[3,4,5]
# LCS=[3,4,5], 길이=3
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
def lcs(a,b):
    m,n=len(a),len(b)
    dp=[[0]*(n+1) for _ in range(m+1)]
    for i in range(1,m+1):
        for j in range(1,n+1):
            if a[i-1]==b[j-1]: dp[i][j]=dp[i-1][j-1]+1
            else: dp[i][j]=max(dp[i-1][j],dp[i][j-1])
    return dp[m][n]
print(lcs([1,3,4,5,2],[3,4,5]))  # 3
</pre>
</details>

### 510 심화 종합 마무리 - 데이터 분석 리포트

딕셔너리, 리스트, 클래스를 모두 활용해 주식 데이터 분석 리포트를 생성하세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class StockReport:
    def __init__(self,symbol,prices):
        self.symbol=symbol; self.prices=prices
    def basic_stats(self):
        return {'최고가':max(self.prices),'최저가':min(self.prices),
                '평균가':round(sum(self.prices)/len(self.prices),2),
                '변동폭':max(self.prices)-min(self.prices)}
    def daily_returns(self):
        return [round((self.prices[i]-self.prices[i-1])/self.prices[i-1]*100,2)
                for i in range(1,len(self.prices))]
    def volatility(self):
        returns=self.daily_returns()
        mean=sum(returns)/len(returns)
        var=sum((r-mean)**2 for r in returns)/len(returns)
        return round(var**0.5,2)
    def trend(self):
        first_half=sum(self.prices[:len(self.prices)//2])
        second_half=sum(self.prices[len(self.prices)//2:])
        return '상승' if second_half>first_half else '하락'
    def report(self):
        print(f'=== {self.symbol} 주가 분석 리포트 ===')
        for k,v in self.basic_stats().items(): print(f'  {k}: {v:,}')
        returns=self.daily_returns()
        print(f'  일평균 수익률: {sum(returns)/len(returns):.2f}%')
        print(f'  변동성(표준편차): {self.volatility()}%')
        print(f'  추세: {self.trend()}')

report=StockReport('005930',[70000,71500,69000,72000,73500,71000,74000,73000,75000,76500])
report.report()
</pre>
