### 106 잘못된 연산자

아래 코드에서 에러가 발생하는 원인을 설명하라.
```python

```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
=> 는 파이썬에서 지원하지 않는 연산자입니다. 크거나 같다는 >= 로 써야 합니다.</br>
<pre>
# 올바른 코드:

</pre>
</details>

````python
# => 대신 >= 를 사용하세요
print(3 >= 4)
````

### 107 if문 기초

### 108 if-else문

아래 코드의 출력 결과를 예상하라.
```python
if 4 < 3:
    print('Hello World.')
else:
    print('Hi, there.')
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
4 < 3이 False이므로 else 블록이 실행됩니다.</br>
<pre>
# 결과: Hi, there.
if 4 < 3:
    print('Hello World.')
else:
    print('Hi, there.')
</pre>
</details>

````python
if 4 < 3:
    print('Hello World.')
else:
    print('Hi, there.')
````

### 109 if문 들여쓰기

아래 코드의 출력 결과를 예상하라.
```python
if True:
    print('1')
    print('2')
else:
    print('3')
print('4')
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
조건이 True이므로 if 블록(1, 2)이 실행됩니다. 4는 if문 밖이므로 항상 출력됩니다.</br>
<pre>
# 결과:
# 1
# 2
# 4
</pre>
</details>

````python
if True:
    print('1')
    print('2')
else:
    print('3')
print('4')
````

### 110 중첩 if문

아래 코드의 출력 결과를 예상하라.
```python
if True:
    if False:
        print('1')
        print('2')
    else:
        print('3')
else:
    print('4')
print('5')
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
바깥 if가 True, 안쪽 if가 False이므로 안쪽 else(3)가 실행됩니다. 5는 항상 출력됩니다.</br>
<pre>
# 결과:
# 3
# 5
</pre>
</details>

````python
if True:
    if False:
        print('1')
        print('2')
    else:
        print('3')
else:
    print('4')
print('5')
````
