### 231 지역 변수 오류 예측

아래 코드를 실행한 결과를 예상하라.
```python
def n_plus_1(n):
    result = n + 1

n_plus_1(3)
print(result)
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
함수 내부에서 선언된 변수(result)는 함수 밖에서 접근할 수 없습니다(지역 변수). 함수 내부 값을 외부로 전달하려면 return을 사용해야 합니다.</br>
<pre>
# 결과: NameError: name 'result' is not defined

# 올바른 방법 (return 사용):
def n_plus_1(n):
    result = n + 1
    return result

result = n_plus_1(3)
print(result)  # 4
</pre>
</details>

````python
# 아래 코드를 실행해서 에러를 확인하고, return을 사용해 수정해보세요
def n_plus_1(n):
    result = n + 1
    return result  # return 추가

result = n_plus_1(3)
print(result)
````

```
실행 예:

```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<pre>

    result = []
    for item in items:

    return result

### 240 삼중 함수 연결 - 결과 예측

아래 코드의 실행 결과를 예측하라.
```python
def 함수0(num):
    return num * 2

def 함수1(num):
    return 함수0(num + 2)

def 함수2(num):
    num = num + 10
    return 함수1(num)

c = 함수2(2)
print(c)
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
함수2(2) → num=2+10=12 → 함수1(12) → 함수0(12+2)=함수0(14) → 14*2=28</br>
<pre>
# 결과: 28
</pre>
</details>

````python
def 함수0(num):
    return num * 2

def 함수1(num):
    return 함수0(num + 2)

def 함수2(num):
    num = num + 10
    return 함수1(num)

c = 함수2(2)
print(c)
````
