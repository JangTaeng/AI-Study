class Portfolio:
    def __init__(self,name): self.name=name; self.holdings={}
    def buy(self,code,qty,price):
        if code in self.holdings:
            old=self.holdings[code]
            total_qty=old['qty']+qty
            avg_price=(old['qty']*old['avg_price']+qty*price)/total_qty
            self.holdings[code]={'qty':total_qty,'avg_price':avg_price}
        else:
            self.holdings[code]={'qty':qty,'avg_price':price}
    def evaluate(self,current_prices):
        total_cost=total_value=0
        for code,h in self.holdings.items():
            cost=h['qty']*h['avg_price']; value=h['qty']*current_prices.get(code,0)
            total_cost+=cost; total_value+=value
            pnl=value-cost; pnl_pct=pnl/cost*100
            print(f'{code}: 평가{value:,}원 손익{pnl:+,}원({pnl_pct:+.1f}%)')
        print(f'총평가: {total_value:,}원 총손익: {total_value-total_cost:+,}원')

class SocialNetwork:
    def __init__(self): self.graph={}
    def add_user(self,user): self.graph.setdefault(user,set())
    def add_friend(self,a,b):
        self.graph.setdefault(a,set()).add(b)
        self.graph.setdefault(b,set()).add(a)
    def friends(self,user): return self.graph.get(user,set())
    def recommend(self,user):
        my_friends=self.friends(user)
        candidates={}
        for f in my_friends:
            for ff in self.friends(f):
                if ff!=user and ff not in my_friends:
                    candidates[ff]=candidates.get(ff,0)+1
        return sorted(candidates.items(),key=lambda x:-x[1])

class Survey:
    def __init__(self,title): self.title=title; self.responses=[]
    def add_response(self,response): self.responses.append(response)
    def count_by(self,field):
        from collections import Counter
        return Counter(r[field] for r in self.responses if field in r)
    def avg(self,field):
        vals=[r[field] for r in self.responses if field in r and isinstance(r[field],(int,float))]
        return sum(vals)/len(vals) if vals else 0
    def report(self):
        print(f'=== {self.title} 결과 ({len(self.responses)}명) ===')
        for field in self.responses[0].keys():
            if isinstance(self.responses[0][field],(int,float)):
                print(f'{field} 평균: {self.avg(field):.1f}')
            else:
                print(f'{field} 분포: {dict(self.count_by(field))}')

class MiniDB:
    def __init__(self): self.tables={}
    def create_table(self,name,schema): self.tables[name]={'schema':schema,'rows':[],'next_id':1}
    def insert(self,table,**data):
        t=self.tables[table]
        row={'id':t['next_id']}
        row.update({k:data.get(k) for k in t['schema']})
        t['rows'].append(row); t['next_id']+=1
        return row['id']
    def select(self,table,where=None):
        rows=self.tables[table]['rows']
        if where: rows=[r for r in rows if all(r.get(k)==v for k,v in where.items())]
        return rows
    def delete(self,table,id_):
        t=self.tables[table]
        t['rows']=[r for r in t['rows'] if r['id']!=id_]

db=MiniDB()
db.create_table('users',['name','age','city'])
