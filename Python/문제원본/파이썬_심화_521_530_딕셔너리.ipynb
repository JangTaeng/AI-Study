### 525 딕셔너리 - 딕셔너리로 LRU 캐시 구현

OrderedDict를 사용해 LRU(Least Recently Used) 캐시를 구현하세요.
```python
cache = LRUCache(capacity=3)
# put/get 연산, 용량 초과 시 가장 오래된 항목 제거
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
from collections import OrderedDict
class LRUCache:
    def __init__(self,capacity):
        self.cache=OrderedDict()
        self.capacity=capacity
    def get(self,key):
        if key not in self.cache: return -1
        self.cache.move_to_end(key)
        return self.cache[key]
    def put(self,key,value):
        if key in self.cache: self.cache.move_to_end(key)
        self.cache[key]=value
        if len(self.cache)>self.capacity:
            self.cache.popitem(last=False)

lru=LRUCache(3)
lru.put('a',1); lru.put('b',2); lru.put('c',3)
print(lru.get('a'))   # 1 (a가 최근으로 이동)
lru.put('d',4)        # b 제거 (가장 오래됨)
print(lru.get('b'))   # -1 (제거됨)
print(dict(lru.cache))
</pre>
</details>

### 526 딕셔너리 - 딕셔너리 직렬화/역직렬화 (커스텀)

딕셔너리를 커스텀 문자열 포맷으로 직렬화하고 역직렬화하세요.
```python
d = {'name':'김파이','age':25}
# 직렬화: 'name=김파이;age=25'
# 역직렬화: {'name':'김파이','age':'25'}
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
d={'name':'김파이','age':25,'city':'서울'}

def serialize(d):
    return ';'.join(f'{k}={v}' for k,v in d.items())

def deserialize(s):
    return dict(item.split('=',1) for item in s.split(';'))

serial=serialize(d)
print('직렬화:', serial)
back=deserialize(serial)
print('역직렬화:', back)
</pre>

### 527 딕셔너리 - 딕셔너리 키 접근 통계

딕셔너리에 접근 횟수를 기록하는 StatDict 클래스를 만드세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
class StatDict(dict):
    def __init__(self,*args,**kwargs):
        super().__init__(*args,**kwargs)
        self._access=dict.fromkeys(self.keys(),0)
    def __getitem__(self,key):
        self._access[key]=self._access.get(key,0)+1
        return super().__getitem__(key)
    def stats(self):
        return sorted(self._access.items(),key=lambda x:-x[1])

sd=StatDict({'a':1,'b':2,'c':3})
sd['a']; sd['a']; sd['b']; sd['a']; sd['c']
print('접근 통계:', sd.stats())
</pre>

### 530 딕셔너리 - 다차원 딕셔너리 평균

```python
temps = {'서울':{'1월':-5,'7월':27},'부산':{'1월':2,'7월':29}}
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
temps={'서울':{'1월':-5,'4월':12,'7월':27,'10월':15},
       '부산':{'1월':2,'4월':15,'7월':29,'10월':18},
       '대구':{'1월':-1,'4월':14,'7월':30,'10월':16}}

print('=== 도시별 연평균 ===')
for city,months in temps.items():
    avg=sum(months.values())/len(months)
    print(f'{city}: {avg:.1f}°C')

print('=== 월별 전국평균 ===')
all_months=list(next(iter(temps.values())).keys())
for month in all_months:
    avg=sum(temps[city][month] for city in temps)/len(temps)
    print(f'{month}: {avg:.1f}°C')
</pre>
