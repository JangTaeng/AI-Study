### 511 딕셔너리 - 두 딕셔너리 교집합/합집합

두 딕셔너리에서 교집합(공통 키+값 동일), 합집합(모든 키), 차집합(한쪽에만 있는 키)을 구하세요.
```python
d1 = {'a':1, 'b':2, 'c':3}
d2 = {'b':2, 'c':99, 'd':4}
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
d1={'a':1,'b':2,'c':3}
d2={'b':2,'c':99,'d':4}

# 교집합: 키와 값이 모두 같은 항목
inter={k:v for k,v in d1.items() if d2.get(k)==v}
print('교집합:', inter)  # {'b': 2}

# 합집합: 두 딕셔너리를 병합 (d2 우선)
union={**d1, **d2}
print('합집합:', union)

# 차집합: d1에만 있는 키
diff={k:v for k,v in d1.items() if k not in d2}
print('차집합(d1-d2):', diff)  # {'a': 1}
</pre>
</details>

### 513 딕셔너리 - 중첩 딕셔너리 병합

두 중첩 딕셔너리를 재귀적으로 병합하세요. 같은 키가 있을 때 d2 값으로 덮어씁니다.
```python
d1 = {'a':{'x':1,'y':2},'b':3}
d2 = {'a':{'y':99,'z':4},'c':5}
# 결과: {'a':{'x':1,'y':99,'z':4},'b':3,'c':5}
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
def deep_merge(d1,d2):
    result=dict(d1)
    for k,v in d2.items():
        if k in result and isinstance(result[k],dict) and isinstance(v,dict):
            result[k]=deep_merge(result[k],v)
        else:
            result[k]=v
    return result

d1={'a':{'x':1,'y':2},'b':3}
d2={'a':{'y':99,'z':4},'c':5}
print(deep_merge(d1,d2))
</pre>

### 516 딕셔너리 - 키 경로로 값 설정

점(.)으로 구분된 경로 문자열로 중첩 딕셔너리의 값을 설정하는 함수를 구현하세요.
```python
d = {}
set_path(d, 'a.b.c', 42)
# 결과: {'a': {'b': {'c': 42}}}
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
def set_path(d, path, value):
    keys=path.split('.')
    for key in keys[:-1]:
        d=d.setdefault(key,{})
    d[keys[-1]]=value

d={}
set_path(d,'a.b.c',42)
set_path(d,'a.b.d',99)
set_path(d,'x.y',10)
print(d)
</pre>

### 518 딕셔너리 - 역방향 인덱스 만들기

<pre>
words=['apple','banana','cherry']
index={}

    for ci,char in enumerate(word):
        index.setdefault(char,[]).append((wi,ci))

for char in ['a','b','e']:
    print(f"'{char}':", index.get(char,[]))

### 519 딕셔너리 - 동적 키 생성 패턴

날짜별 판매 데이터를 딕셔너리로 집계하고, 없는 날짜는 자동으로 0으로 초기화되도록 구현하세요.

```python
sales = [('2024-01',100),('2024-02',200),('2024-01',150),('2024-03',80)]

```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
from collections import defaultdict
sales=[('2024-01',100),('2024-02',200),('2024-01',150),('2024-03',80)]
monthly=defaultdict(int)
for date,amount in sales:
    monthly[date]+=amount
for date,total in sorted(monthly.items()):
    print(f'{date}: {total:,}원')

### 520 딕셔너리 - 딕셔너리로 그래프 최단 경로

딕셔너리로 표현된 가중치 그래프에서 다익스트라 알고리즘으로 최단 거리를 구하세요.
```python
graph = {'A':{'B':1,'C':4},'B':{'C':2,'D':5},'C':{'D':1},'D':{}}
# A에서 각 노드까지 최단 거리
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
import heapq
graph={'A':{'B':1,'C':4},'B':{'C':2,'D':5},'C':{'D':1},'D':{}}
def dijkstra(graph,start):
    dist={node:float('inf') for node in graph}
    dist[start]=0
    pq=[(0,start)]
    while pq:
        d,u=heapq.heappop(pq)
        if d>dist[u]: continue
        for v,w in graph[u].items():
            if dist[u]+w<dist[v]:
                dist[v]=dist[u]+w
                heapq.heappush(pq,(dist[v],v))
    return dist
print(dijkstra(graph,'A'))
</pre>
</details>
