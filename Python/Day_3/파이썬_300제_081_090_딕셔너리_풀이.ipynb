### 081 별 표현식 (좌측 8개)

star expression(*)을 사용하여 좌측 8개의 값을 valid_score 변수에 바인딩하세요.
```python
scores = [8.8, 8.9, 8.7, 9.2, 9.3, 9.7, 9.9, 9.5, 7.8, 9.4]
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<code>*변수</code>는 나머지 모든 값을 리스트로 받습니다. <code>_</code>는 필요 없는 값을 버릴 때 관례적으로 사용합니다.</br>
<pre>
scores = [8.8, 8.9, 8.7, 9.2, 9.3, 9.7, 9.9, 9.5, 7.8, 9.4]
*valid_score, _, _ = scores
print(valid_score)
</pre>
</details>

````python
scores = [8.8, 8.9, 8.7, 9.2, 9.3, 9.7, 9.9, 9.5, 7.8, 9.4]
# 여기에 코드를 작성하세요
*valid_score, _, _ = scores
# 언팩킹(Unpacking) 기능을 사용
# *변수명 은 "나머지 전부를 여기에 담아줘" 라는 의미
# _ 는 "이 값은 필요 없어, 버릴 거야" 라는 관례적 표현
print(valid_score)
````

### 082 별 표현식 (우측 8개)

star expression(*)을 사용하여 우측 8개의 값을 valid_score 변수에 바인딩하세요.
```python
scores = [8.8, 8.9, 8.7, 9.2, 9.3, 9.7, 9.9, 9.5, 7.8, 9.4]
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
앞의 두 값을 버리고, 나머지 값들을 <code>*valid_score</code>로 받습니다.</br>
<pre>
scores = [8.8, 8.9, 8.7, 9.2, 9.3, 9.7, 9.9, 9.5, 7.8, 9.4]
a, b, *valid_score = scores
print(valid_score)
</pre>
</details>

````python
scores = [8.8, 8.9, 8.7, 9.2, 9.3, 9.7, 9.9, 9.5, 7.8, 9.4]
# 여기에 코드를 작성하세요
_,_,*valid_score = scores
print(valid_score)
````

### 090 존재하지 않는 키 접근

다음 코드에서 에러가 발생한 원인을 설명하세요.
```python
icecream = {'폴라포': 1200, '빵빠레': 1800, '월드콘': 1500, '메로나': 1000}
icecream['누가바']
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
딕셔너리에 없는 키로 접근하면 <code>KeyError</code>가 발생합니다. <code>get()</code> 메서드를 사용하면 키가 없을 때 에러 대신 None을 반환합니다.</br>
<pre>
icecream = {'폴라포': 1200, '빵빠레': 1800, '월드콘': 1500, '메로나': 1000}

# 에러 발생:
# icecream['누가바']  # KeyError

# 안전한 방법:
print(icecream.get('누가바'))        # None
print(icecream.get('누가바', 0))     # 0 (기본값 지정)
</pre>
</details>

````python
icecream = {'폴라포': 1200, '빵빠레': 1800, '월드콘': 1500, '메로나': 1000}
# 직접 실행해서 에러를 확인해보세요

print(icecream.get('누가바'))        # None
print(icecream.get('누가바', 0))     # 0 (기본값 지정)
print(icecream)

icecream['누가바']
#KeyError: '누가바'
````
