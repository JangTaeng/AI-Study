### 061 리스트 슬라이싱

### 066 join 메서드 (공백 구분)

join() 메서드를 사용해서 리스트를 공백으로 구분하여 출력하세요.
```python
interest = ['삼성전자', 'LG전자', 'Naver', 'SK하이닉스', '미래에셋대우']

```

<details>

### 069 split → 리스트 변환

회사 이름이 '/'로 구분된 문자열을 interest 이름의 리스트로 분리하세요.
```python
string = "삼성전자/LG전자/Naver"
```
```
실행 예시:
['삼성전자', 'LG전자', 'Naver']
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<code>split(구분자)</code>는 문자열을 구분자 기준으로 나눠 리스트로 반환합니다.</br>
<pre>
string = "삼성전자/LG전자/Naver"
interest = string.split("/")
print(interest)
</pre>
</details>

````python
string = "삼성전자/LG전자/Naver"
# 여기에 코드를 작성하세요
print(string.split("/"))
````

### 070 리스트 정렬

리스트의 값을 오름차순으로 정렬하세요.
```python
data = [2, 4, 3, 1, 5, 10, 9]
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<code>sort()</code>는 리스트를 제자리에서 정렬합니다. <code>sorted()</code>는 새 리스트를 반환합니다.</br>
<pre>
data = [2, 4, 3, 1, 5, 10, 9]
data.sort()
print(data)

# 또는 sorted() 사용:
data2 = [2, 4, 3, 1, 5, 10, 9]
print(sorted(data2))
</pre>
</details>

````python
data = [2, 4, 3, 1, 5, 10, 9]
# 여기에 코드를 작성하세요
data.sort()
print(data)
````
