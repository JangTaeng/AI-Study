### 074 튜플은 immutable

다음 코드를 실행해보고 오류가 발생하는 원인을 설명하세요.
```python
t = (1, 2, 3)
t[0] = 'a'
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
튜플은 한 번 만들어지면 원소의 값을 변경할 수 없습니다(immutable). 변경하려 하면 TypeError가 발생합니다.</br>
<pre>
t = (1, 2, 3)
t[0] = 'a'  # TypeError: 'tuple' object does not support item assignment
</pre>
</details>

````python
# 직접 실행해서 오류를 확인해보세요
t = (1, 2, 3)
t[0] = 'a'

#TypeError: 'tuple' object does not support item assignment
# 튜플은 한 번 만들어지면 원소의 값을 변경할 수 없습니다
````

### 076 튜플 값 변경하기


### 079 튜플 언패킹

다음 코드의 실행 결과를 예상하세요.
```python
temp = ('apple', 'banana', 'cake')
a, b, c = temp
print(a, b, c)
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
튜플 언패킹은 튜플의 각 원소를 여러 변수에 동시에 할당합니다.</br>
<pre>
temp = ('apple', 'banana', 'cake')
a, b, c = temp
print(a, b, c)  # apple banana cake
</pre>
</details>

````python
# 직접 실행해서 결과를 확인해보세요
temp = ('apple', 'banana', 'cake')
a, b, c = temp
print(a, b, c)
# a = apple, b = banana, c = cake
````
