### 041 upper 메서드

다음 문자열을 대문자 BTC_KRW로 변경하세요.
```python

```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<code>upper()</code> 메서드는 문자열을 모두 대문자로 변환한 새 문자열을 반환합니다.</br>
<pre>

</pre>
</details>

````python
ticker = "btc_krw"
# 여기에 코드를 작성하세요
print(ticker.upper())
````

### 044 endswith 메서드

파일 이름이 'xlsx'로 끝나는지 확인해보세요.
```python
file_name = "보고서.xlsx"
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<code>endswith()</code>는 문자열이 특정 문자열로 끝나는지 True/False로 반환합니다.</br>
<pre>
file_name = "보고서.xlsx"
result = file_name.endswith("xlsx")
print(result)  # True
</pre>
</details>

````python
file_name = "보고서.xlsx"
# 여기에 코드를 작성하세요
print(file_name.endswith("xlsx"))
````

### 045 endswith 메서드

파일 이름이 'xlsx' 또는 'xls'로 끝나는지 확인해보세요.

```python
file_name = "보고서.xlsx"
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<code>endswith()</code>에 튜플을 넣으면 여러 확장자를 한 번에 확인할 수 있습니다.</br>
<pre>
file_name = "보고서.xlsx"

</pre>
</details>

````python
file_name = "보고서.xlsx"
# 여기에 코드를 작성하세요
print(file_name.endswith(("xlsx", "xls")))
````

### 047 split 메서드

공백을 기준으로 문자열을 나눠보세요.
```python
a = "hello world"
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<code>split()</code>은 인자 없이 호출하면 공백을 기준으로 문자열을 분리해 리스트로 반환합니다.</br>
<pre>
a = "hello world"
result = a.split()
print(result)  # ['hello', 'world']
</pre>
</details>

````python
a = "hello world"
# 여기에 코드를 작성하세요
print(a.split())
````

### 048 split 메서드


<details>

### 050 rstrip 메서드

문자열 오른쪽의 공백을 제거해보세요.
```python
data = "039490     "
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<code>rstrip()</code>은 오른쪽 공백만 제거합니다. <code>lstrip()</code>은 왼쪽, <code>strip()</code>은 양쪽 모두 제거합니다.</br>
<pre>
data = "039490     "
data = data.rstrip()
print(data)  # '039490'
</pre>
</details>

````python
data = "039490     "
# 여기에 코드를 작성하세요
print(data.rstrip())
````
