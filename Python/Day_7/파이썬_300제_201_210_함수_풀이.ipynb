### 205 함수 정의 전 호출 오류

아래 코드에서 에러가 발생하는 이유를 설명하라.
```python
hello()
def hello():
    print("Hi")
```
```
실행 예:
NameError: name 'hello' is not defined
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
파이썬은 위에서 아래로 순서대로 실행됩니다. 함수가 정의되기 전에 호출하면 NameError가 발생합니다. 함수는 항상 호출 전에 정의되어 있어야 합니다.</br>

    print("Hi")
hello()  # 정의 후 호출

</details>

### 207 함수와 일반 코드 실행 순서

```python

```

### 208 복잡한 실행 순서 예측

### 209 함수 안에서 함수 호출

아래 코드의 실행 결과를 예측하라.
```python
def message1():
    print("A")

def message2():
    print("B")
    message1()

message2()
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
message2() 호출 → "B" 출력 → message1() 호출 → "A" 출력</br>
<pre>
# 결과:
# B
# A
</pre>
</details>

````python
def message1():
    print("A")

def message2():
    print("B")
    message1()

message2()

# B A
````

### 210 복합 함수 실행 순서

아래 코드의 실행 결과를 예측하라.
```python
def message1():
    print("A")

def message2():
    print("B")

def message3():
    for i in range(3):
        message2()
        print("C")
    message1()

message3()
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
message3() → for문 3번 (message2() → "B", print("C")) → message1() → "A"</br>
<pre>
# 결과:
# B
# C
# B
# C
# B
# C
# A
</pre>
</details>

````python
def message1():
    print("A")

def message2():
    print("B")

def message3():
    for i in range(3):
        message2()
        print("C")
    message1()

message3()

# BCBCBCA
````
