### 191 중첩 for문 - 수수료 포함 가격 출력

data에는 종목들의 OHLC(open/high/low/close) 가격 정보가 저장되어 있다. 수수료를 0.014%로 가정할 때, 각 가격에 수수료를 포함한 가격을 한 라인에 하나씩 출력하라.
```python
data = [
    [ 2000,  3050,  2050,  1980],
    [ 7500,  2050,  2050,  1980],
    [15450, 15050, 15550, 14900]
]
```
```
실행 예:
2000.28
3050.427
...
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
중첩 for문으로 2차원 리스트를 순회합니다.</br>
<pre>
data = [
    [ 2000,  3050,  2050,  1980],
    [ 7500,  2050,  2050,  1980],
    [15450, 15050, 15550, 14900]
]
for line in data:
    for column in line:

</pre>
</details>

````python
data = [
    [ 2000,  3050,  2050,  1980],
    [ 7500,  2050,  2050,  1980],
    [15450, 15050, 15550, 14900]
]
# 여기에 코드를 작성하세요
for line in data:
    for num in line:
      print(num * 1.00014)
````

### 192 중첩 for문 - 행 구분자 추가

191번의 출력 결과에 행단위로 '----' 구분자를 추가하라.
```
실행 예:

7501.05
...
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
안쪽 for문이 끝날 때마다(바깥 for문 들여쓰기에서) 구분자를 출력합니다.</br>
<pre>
for line in data:
    for column in line:
        print(column * 1.00014)
    print("----")
</pre>
</details>

````python
data = [
    [ 2000,  3050,  2050,  1980],
    [ 7500,  2050,  2050,  1980],
    [15450, 15050, 15550, 14900]
]
# 여기에 코드를 작성하세요
for line in data:
  for num in line:
    print(num * 1.00014)
  print("----")
````

### 193 중첩 for문 - 1차원 리스트로 저장

191번 문제의 결괏값을 result 이름의 리스트에 1차원 배열로 저장하라.
```
실행 예:
>> print(result)
[2000.28, 3050.427, 2050.287, 1980.277, 7501.05, ...]
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
빈 리스트를 만들고 append()로 값을 추가합니다.</br>
<pre>
result = []
for line in data:
    for column in line:
        result.append(column * 1.00014)
print(result)
</pre>
</details>

````python
data = [
    [ 2000,  3050,  2050,  1980],
    [ 7500,  2050,  2050,  1980],
    [15450, 15050, 15550, 14900]
]
# 여기에 코드를 작성하세요
result = []
for line in data:
  for num in line:
    result.append(num * 1.00014)
print(result)
````

### 194 중첩 for문 - 2차원 리스트로 저장

191번 문제의 결괏값을 result 이름의 리스트에 2차원 배열로 저장하라.
```
실행 예:
>> print(result)
[[2000.28, 3050.427, 2050.287, 1980.277], [7501.05, ...], [...]]
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
행마다 sub 리스트를 만들어 값을 추가하고, sub를 result에 추가합니다.</br>
<pre>
result = []
for line in data:
    sub = []
    for column in line:
        sub.append(column * 1.00014)
    result.append(sub)
print(result)
</pre>
</details>

````python
data = [
    [ 2000,  3050,  2050,  1980],
    [ 7500,  2050,  2050,  1980],
    [15450, 15050, 15550, 14900]
]
# 여기에 코드를 작성하세요
result = []
for line in data:
    sub = []
    for column in line:
        sub.append(column * 1.00014)
    result.append(sub)
print(result)
````

### 195 2차원 리스트 - 종가 출력

ohlc 리스트에는 시가(open), 고가(high), 저가(low), 종가(close)가 날짜별로 저장돼 있다. 종가 데이터만 출력하라.

```
실행 예:
```

<pre>

        [100, 110, 70, 100],
        [200, 210, 180, 190],
        [300, 310, 300, 310]]
for row in ohlc[1:]:
    print(row[3])
</pre>
</details>

````python
ohlc = [["open", "high", "low", "close"],
        [100, 110, 70, 100],
        [200, 210, 180, 190],
        [300, 310, 300, 310]]
# 여기에 코드를 작성하세요

for line in ohlc[1:]:
````
````python
    print(line[3])
````

### 198 2차원 리스트 - 변동폭 리스트 저장

ohlc 리스트에서 고가와 저가의 차이(변동폭)를 volatility 리스트에 저장하라.
```python
ohlc = [["open", "high", "low", "close"],
        [100, 110, 70, 100],
        [200, 210, 180, 190],
        [300, 310, 300, 310]]
```
```
실행 예:
>> print(volatility)
[40, 30, 10]
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
고가(인덱스 1)에서 저가(인덱스 2)를 뺀 값을 리스트에 추가합니다.</br>
<pre>
volatility = []
for row in ohlc[1:]:

print(volatility)
</pre>
</details>

````python
ohlc = [["open", "high", "low", "close"],
        [100, 110, 70, 100],
        [200, 210, 180, 190],
        [300, 310, 300, 310]]
# 여기에 코드를 작성하세요
volatility = []
for line in ohlc[1:]:
    volatility.append(line[1] - line[2])
print(volatility)
````

### 200 2차원 리스트 - 총 수익금 계산

ohlc 리스트에서 시가에 매수해서 종가에 매도했을 경우 총 수익금을 계산하라.
```python
ohlc = [["open", "high", "low", "close"],
        [100, 110, 70, 100],
        [200, 210, 180, 190],
        [300, 310, 300, 310]]
```
- 1일차: 100 - 100 = 0원
- 2일차: 190 - 200 = -10원
- 3일차: 310 - 300 = 10원
```
실행 예:
0
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<pre>
ohlc = [["open", "high", "low", "close"],
        [100, 110, 70, 100],
        [200, 210, 180, 190],
        [300, 310, 300, 310]]
profit = 0
for row in ohlc[1:]:
    profit += (row[3] - row[0])
print(profit)
</pre>
</details>

````python
ohlc = [["open", "high", "low", "close"],
        [100, 110, 70, 100],
        [200, 210, 180, 190],
        [300, 310, 300, 310]]
# 여기에 코드를 작성하세요
total = 0;
for line in ohlc[1:]:
  total += (line[3] - line[0])
print(total)
````
