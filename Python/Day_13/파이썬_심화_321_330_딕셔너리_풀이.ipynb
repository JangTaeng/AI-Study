### 326 메모이제이션으로 피보나치

딕셔너리 캐시로 피보나치를 효율적으로 계산하세요.

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
cache={}
def fib(n):
    if n in cache: return cache[n]
    if n<=1: return n
    cache[n]=fib(n-1)+fib(n-2)
    return cache[n]
for i in range(11):
    print(f'fib({i})={fib(i)}')
</pre>
</details>

````python
# 여기에 코드를 작성하세요
cache={}
def fib(n):
    if n in cache: return cache[n]
    if n<=1: return n
    cache[n]=fib(n-1)+fib(n-2)
    return cache[n]
for i in range(11):
    print(f'fib({i})={fib(i)}')
````

### 330 딕셔너리 순회 4가지

딕셔너리를 순회하는 4가지 방법을 모두 사용하세요.
```python
scores={'민수':85,'지영':62,'철수':91}
```

<details>
<summary>정답 확인 (클릭!)</summary>
<pre>
scores={'민수':85,'지영':62,'철수':91}
print('--- 키만 ---')
for k in scores: print(k)
print('--- 값만 ---')
for v in scores.values(): print(v)
print('--- 키+값 ---')
for k,v in scores.items(): print(k,v)
print('--- enumerate ---')
for i,(k,v) in enumerate(scores.items()): print(i,k,v)
</pre>
</details>

````python
scores={'민수':85,'지영':62,'철수':91}
# 여기에 코드를 작성하세요
print()
print('[키만]')
for k in scores: print(k)
print()
print('[값만]')
for v in scores.values(): print(v)
print()
print('[키+값]')
for k,v in scores.items(): print(k,v)
print()
print('[enumerate]')
for i,(k,v) in enumerate(scores.items()): print(i,k,v)
````
