
### 027 문자열 다루기

url에 저장된 웹 페이지 주소에서 도메인을 출력하세요.
```python
url = "http://sharebook.kr"
```
```
실행 예:
kr
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
split() 메서드는 특정 문자를 기준으로 문자열을 나눠 리스트로 반환합니다. 마지막 요소를 인덱싱하면 도메인만 얻을 수 있습니다.</br>

</pre>
</details>

### 028 문자열은 immutable

아래 코드의 실행 결과를 예상해보세요.
```python
lang = 'python'
lang[0] = 'P'
print(lang)
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
문자열은 수정할 수 없는(immutable) 자료형입니다. 인덱싱으로 값을 바꾸려 하면 TypeError가 발생합니다.</br>
<pre>
# 실행 결과:
# TypeError: 'str' object does not support item assignment
lang = 'python'
lang[0] = 'P'  # 오류 발생!
print(lang)
</pre>
</details>

````python
# 직접 실행해서 어떤 오류가 나는지 확인해보세요
lang = 'python'
lang[0] = 'P'
print(lang)
# TypeError: 'str' object does not support item assignment
````

### 029 replace 메서드

아래 문자열에서 소문자 'a'를 대문자 'A'로 변경하세요.
```python
string = 'abcdfe2a354a32a'
```
```
실행 예:
Abcdfe2A354A32A
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>

string = string.replace('a', 'A')

### 030 replace 메서드

아래 코드의 실행 결과를 예상해보세요.
```python
string = 'abcd'
string.replace('b', 'B')
print(string)
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<code>abcd</code>가 그대로 출력됩니다. 문자열은 변경할 수 없는 자료형이므로 replace 메서드는 원본을 그대로 두고 변경된 새로운 문자열을 리턴합니다. 반환값을 변수에 저장해야 사용할 수 있습니다.</br>
<pre>
string = 'abcd'
string.replace('b', 'B')  # 반환값을 저장하지 않음
print(string)  # abcd (원본 그대로)

# 올바른 방법:
string = string.replace('b', 'B')
print(string)  # aBcd
</pre>
</details>

````python
# 직접 실행해서 결과를 확인해보세요
string = 'abcd'
string.replace('b', 'B') # 반환값을 변수에 저장해야 적용됨 또는 직접 프린트
print(string)
````
