### 164 range - 99부터 0까지 역순 출력

99부터 0까지 1씩 감소하는 숫자들을 한 라인에 하나씩 출력하라.

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
뺄셈을 사용해 역순으로 출력합니다.</br>
<pre>
for i in range(100):
    print(99 - i)

# 또는 range의 역순:

</pre>
</details>

````python
# 여기에 코드를 작성하세요
for i in range(100):
    print(99 - i)
````

### 167 구구단 3단 (홀수 번째만)

```

```

<details>

num = 3

### 168 1~10 합계

1~10까지의 숫자를 모두 더한 값을 출력하는 프로그램을 for문을 사용하여 작성하라.
```
실행 예:
합 : 55
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
누적 합산 변수(hab)를 0으로 초기화하고 for문으로 더해나갑니다.</br>
<pre>
hab = 0
for i in range(1, 11):
    hab += i
print("합 :", hab)
</pre>
</details>

````python
# 여기에 코드를 작성하세요
hab = 0
for i in range(1, 11):
    hab += i
print("합 :", hab)
````

### 170 1~10 곱셈 (팩토리얼)

1~10까지의 숫자를 모두 곱한 값을 출력하는 프로그램을 for문을 사용하여 작성하라.

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
누적 곱셈 변수(result)를 1로 초기화합니다. (0으로 초기화하면 항상 0이 됨)</br>
<pre>
result = 1
for i in range(1, 11):
    result *= i
print(result)  # 3628800
</pre>
</details>

````python
# 여기에 코드를 작성하세요
result = 1
for i in range(1, 11):
    result *= i
print(result)
````
