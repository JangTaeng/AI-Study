### 172 enumerate - 인덱스와 값 함께 출력

아래와 같이 인덱스와 데이터를 함께 출력하라.

```
실행 예:

2 32000
3 32500
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
enumerate()는 인덱스와 값을 동시에 반환합니다.</br>
<pre>

</pre>
</details>

````python
price_list = [32100, 32150, 32000, 32500]
# 여기에 코드를 작성하세요
for i, data in enumerate(price_list):
    print(i, data)
````

### 179 3일 이동평균 계산

리스트에는 6일 간의 종가 데이터가 저장되어 있다. 종가 데이터의 3일 이동 평균을 계산하고 출력하라.

<pre>
my_list = [100, 200, 400, 800, 1000, 1300]
for i in range(1, len(my_list) - 1):

### 180 변동폭 리스트 저장

리스트에 5일간의 저가, 고가 정보가 저장돼 있다. 고가와 저가의 차를 변동폭으로 정의할 때, 5일간의 변동폭을 volatility 리스트에 저장하라.
```python
low_prices  = [100, 200, 400, 800, 1000]
high_prices = [150, 300, 430, 880, 1000]
```
```
실행 예:
>> print(volatility)
[50, 100, 30, 80, 0]
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
빈 리스트를 만들고 append()로 차분값을 추가합니다.</br>
<pre>
low_prices  = [100, 200, 400, 800, 1000]
high_prices = [150, 300, 430, 880, 1000]
volatility = []
for i in range(len(low_prices)):
    volatility.append(high_prices[i] - low_prices[i])
print(volatility)
</pre>
</details>

````python
low_prices  = [100, 200, 400, 800, 1000]
high_prices = [150, 300, 430, 880, 1000]
# 여기에 코드를 작성하세요
volatility = []
for i in range(len(low_prices)):
    volatility.append(high_prices[i] - low_prices[i])
print(volatility)
````
