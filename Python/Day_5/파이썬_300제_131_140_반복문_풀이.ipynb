### 131 for문 기초 - 결과 예측

```python

```

<details>

<pre>

# 귤
# 수박
</pre>
</details>

````python
과일 = ["사과", "귤", "수박"]
for 변수 in 과일:
    print(변수)
````

### 136 코드를 for문으로 변환

다음 코드를 for문으로 작성하라.
```python
변수 = 10
print(변수)
변수 = 20
print(변수)
변수 = 30
print(변수)
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
변하는 값(10, 20, 30)은 리스트로, 변하지 않는 코드(print)는 들여쓰기로 작성합니다.</br>
<pre>
for 변수 in [10, 20, 30]:
    print(변수)
</pre>
</details>

````python
# 아래 코드를 for문으로 변환해보세요
변수 = 10
print(변수)
변수 = 20
print(변수)
변수 = 30
print(변수)
````


<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>

<pre>
print("++++")
for 변수 in [10, 20, 30]:
    print(변수)
</pre>
</details>

````python
# 아래 코드를 for문으로 변환해보세요
print("++++")
print(10)
print(20)
print(30)
````

### 140 for문으로 변환 (변하는 값 없음)

다음 코드를 for문으로 작성하라.
```python
print("-------")
print("-------")
print("-------")
print("-------")
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
변하는 값이 없어도, 데이터가 4개인 리스트로 4번 반복할 수 있습니다.</br>
<pre>
for 변수 in [1, 2, 3, 4]:
    print("-------")

# 또는 range 활용:
for i in range(4):
    print("-------")
</pre>
</details>

````python
# 아래 코드를 for문으로 변환해보세요
print("-------")
print("-------")
print("-------")
print("-------")
````
