### 142 for문 + 문자열 출력

for문을 사용해서 리스트에 저장된 값을 다음과 같이 출력하라.
```python
리스트 = ["김밥", "라면", "튀김"]
```
```
실행 예:
오늘의 메뉴: 김밥
오늘의 메뉴: 라면
오늘의 메뉴: 튀김
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<pre>
리스트 = ["김밥", "라면", "튀김"]
for 메뉴 in 리스트:
    print("오늘의 메뉴: " + 메뉴)

# 또는:
for 메뉴 in 리스트:
    print("오늘의 메뉴:", 메뉴)
</pre>
</details>

````python
리스트 = ["김밥", "라면", "튀김"]
# 여기에 코드를 작성하세요
````

### 147 for문 + 곱셈 출력

for문을 사용해서 다음과 같이 출력하라.
```python
리스트 = [1, 2, 3]
```
```
실행 예:
3 x 1 = 3
3 x 2 = 6
3 x 3 = 9
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<pre>
리스트 = [1, 2, 3]
for 변수 in 리스트:
    print("3 x ", 변수, "=", 3 * 변수)

# format 활용:
for 변수 in 리스트:
    print("3 x {} = {}".format(변수, 3 * 변수))
</pre>
</details>

````python
리스트 = [1, 2, 3]
# 여기에 코드를 작성하세요
````
