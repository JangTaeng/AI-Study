### 219 사칙연산 함수

두 개의 숫자를 입력받아 합/차/곱/나눗셈을 출력하는 print_arithmetic_operation 함수를 작성하라.
```
실행 예:
print_arithmetic_operation(3, 4)
3 + 4 = 7
3 - 4 = -1
3 * 4 = 12
3 / 4 = 0.75
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<pre>
def print_arithmetic_operation(a, b):
    print(a, "+", b, "=", a + b)
    print(a, "-", b, "=", a - b)
    print(a, "*", b, "=", a * b)
    print(a, "/", b, "=", a / b)

print_arithmetic_operation(3, 4)
</pre>
</details>

````python
# 여기에 코드를 작성하세요
def print_arithmetic_operation(a,b):
  print(a,"+",b,"=",a+b)
  print(a,"-",b,"=",a-b)
  print(a,"*",b,"=",a*b)
  print(a,"/",b,"=",a/b)

print_arithmetic_operation(3,4)
````

### 220 최댓값 출력 함수

세 개의 숫자를 입력받아 가장 큰 수를 출력하는 print_max 함수를 정의하라. 단 if문을 사용해서 비교하라.
```
실행 예:
print_max(3, 7, 5)
7
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<pre>
def print_max(a, b, c):
    max_val = 0
    if a > max_val:
        max_val = a
    if b > max_val:
        max_val = b
    if c > max_val:
        max_val = c
    print(max_val)

# 또는 내장함수 활용:
def print_max2(a, b, c):
    print(max(a, b, c))
</pre>
</details>

````python
# 여기에 코드를 작성하세요
def print_max(a,b,c):
  max = 0
  if(a>max):
    max = a
  if(b>max):
    max = b
  if(c>max):
    max = c
  print(max)

print_max(3,7,5)
````
