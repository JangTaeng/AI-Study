### 257 클래스 메서드 - who()

사람(Human) 클래스에서 이름, 나이, 성별을 출력하는 who() 메서드를 추가하세요.
```
실행 예:
>> areum.who()
이름: 아름 나이: 25 성별: 여자
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
메서드의 첫 번째 파라미터는 항상 self(인스턴스 자신)입니다.</br>
<pre>
class Human:
    def __init__(self, name, age, sex):
        self.name = name
        self.age = age
        self.sex = sex

    def who(self):
        print("이름: {} 나이: {} 성별: {}".format(self.name, self.age, self.sex))

areum = Human("아름", 25, "여자")
areum.who()
</pre>
</details>

````python
# 여기에 코드를 작성하세요
class Human:
  def __init__(self, name, age, sex):
    self.name = name
    self.age = age
    self.sex = sex

  def who(self):
    print("이름: {} 나이: {} 성별: {}".format(self.name, self.age, self.sex))

areum = Human("아름", 25, "여자")
areum.who()
````

### 258 클래스 메서드 - setInfo()

사람(Human) 클래스에 (이름, 나이, 성별)을 받는 setInfo 메서드를 추가하세요.
```
실행 예:
>> areum = Human("모름", 0, "모름")
>> areum.setInfo("아름", 25, "여자")
>> areum.who()
이름: 아름 나이: 25 성별: 여자
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
self.속성 = 새값 으로 속성을 업데이트합니다.</br>
<pre>
class Human:
    def __init__(self, name, age, sex):
        self.name = name
        self.age = age
        self.sex = sex

    def who(self):
        print("이름: {} 나이: {} 성별: {}".format(self.name, self.age, self.sex))

    def setInfo(self, name, age, sex):
        self.name = name
        self.age = age
        self.sex = sex

areum = Human("불명", "미상", "모름")
areum.who()
areum.setInfo("아름", 25, "여자")
areum.who()
</pre>
</details>

````python
# 여기에 코드를 작성하세요
class Human:
  def __init__(self, name, age, sex):
    self.name = name
    self.age = age
    self.sex = sex

  def who(self):
    print("이름: {} 나이: {} 성별: {}".format(self.name, self.age, self.sex))

  def setInfo(self, name, age, sex):
    self.name = name
    self.age = age
    self.sex = sex

areum = Human("불명", "미상", "모름")
areum.who()

areum.setInfo("아름", 25, "여자")
areum.who()

areum.setInfo("찰스", 23, "남자")
areum.who()
````

### 259 클래스 소멸자

사람(Human) 클래스에 '나의 죽음을 알리지 말라'를 출력하는 소멸자를 추가하세요.
```
실행 예:
>> areum = Human("아름", 25, "여자")
>> del areum
나의 죽음을 알리지 말라
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
__del__은 인스턴스가 삭제될 때 자동으로 호출되는 소멸자 메서드입니다.</br>
<pre>
class Human:
    def __init__(self, name, age, sex):
        self.name = name
        self.age = age
        self.sex = sex

    def __del__(self):
        print("나의 죽음을 알리지말라")

areum = Human("아름", 25, "여자")
del(areum)
</pre>
</details>

````python
# 여기에 코드를 작성하세요
class Human:
  def __init__(self, name, age, sex):
    self.name = name
    self.age = age
    self.sex = sex

  def __del__(self):
    print("나의 죽음을 알리지말라")

areum = Human("아름", 25, "여자")
del(areum)
````

### 260 클래스 메서드 오류 원인

아래와 같은 에러가 발생한 원인에 대해 설명하세요.
```python
class OMG:
    def print():
        print("Oh my god")

myStock = OMG()
myStock.print()
# TypeError: print() takes 0 positional arguments but 1 was given
```

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
클래스 메서드는 반드시 첫 번째 파라미터로 self(인스턴스 자신)를 받아야 합니다. myStock.print() 호출 시 파이썬이 자동으로 인스턴스(myStock)를 첫 번째 인자로 전달하는데, self 파라미터가 없어서 에러가 발생합니다.</br>
<pre>
# 올바른 코드:
class OMG:
    def print(self):  # self 추가
        print("Oh my god")

myStock = OMG()
myStock.print()
</pre>
</details>

````python
# 올바르게 수정해보세요
class OMG:
    def print(self):  # self를 추가해야 합니다 / 클래스 메서드는 반드시 첫 번째 파라미터로 self(인스턴스 자신)를 받아야 합니다
        print("Oh my god")

myStock = OMG()
myStock.print()
````
