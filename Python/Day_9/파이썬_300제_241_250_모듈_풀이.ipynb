### 243 timedelta - 과거 날짜 출력

datetime 모듈의 timedelta를 사용해서 오늘로부터 5일, 4일, 3일, 2일, 1일 전의 날짜를 화면에 출력해보세요.

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
timedelta(days=n)으로 날짜 차이를 만들고 현재 시간에서 뺍니다.</br>
<pre>
import datetime

now = datetime.datetime.now()
for day in range(5, 0, -1):
    delta = datetime.timedelta(days=day)
    date = now - delta
    print(date)
</pre>
</details>

````python
import datetime

# 여기에 코드를 작성하세요
now = datetime.datetime.now()
for day in range(5, 0, -1):
    delta = datetime.timedelta(days=day)
    date = now - delta
    print(date)
````

### 247 모듈 임포트 4가지 방식

모듈을 임포트하는 4가지 방식에 대해 설명해보세요.

<details>
<summary>여기를 눌러서 정답을 확인하세요 (클릭!)</summary>
<pre>
# 1) import 모듈명
import datetime
datetime.datetime.now()

# 2) import 모듈명 as 별칭
import datetime as dt
dt.datetime.now()

# 3) from 모듈명 import 특정함수
from datetime import datetime
datetime.now()

# 4) from 모듈명 import *
from datetime import *
datetime.now()
</pre>
</details>

````python
# 4가지 방식을 직접 써보세요

# 1) import 모듈명
import datetime
print(datetime.datetime.now())

# 2) import 모듈명 as 별칭
import datetime as dt
print(dt.datetime.now())

# 3) from 모듈명 import 특정함수
from datetime import datetime
print(datetime.now())

# 4) from 모듈명 import 모든함수
from datetime import *
datetime.now()
````
